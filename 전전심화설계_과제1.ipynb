torch.cat(tensors, dim=0, out=None)
tensors: 연결할 텐서들의 시퀀스(리스트 등)입니다.
dim (옵션): 연결할 때 사용할 차원(축)을 지정합니다. 이 매개변수를 사용하여 연결할 차원을 선택할 수 있으며, 기본값은 0입니다.
out (옵션): 결과를 저장할 출력 텐서를 지정합니다. 이 매개변수를 사용하면 새로운 텐서를 생성하지 않고 결과를 저장할 수 있습니다.
예를 들어, 두 개의 텐서를 연결하려면 다음과 같이 사용할 수 있습니다:

torch.view는 모양만 바꿔주고 메모리는 똑같이 공유한다.
torch.shape는 모양만 바꿔주는 것이 아니라 메모리도 다른 공간에 할당되어 다른 텐서를 만들어 준다

>> x = torch.randn(2, 3, 5)
>>> x.size()
torch.Size([2, 3, 5])
>>> torch.permute(x, (2, 0, 1)).size()
torch.Size([5, 2, 3])
torch.permute는 차원의 순서를 서로 바꿔주고 기존의 tensor와 data를 공유한다.

torch.transpose는 2개의 차원을 transpose하고 기존의 tensor와 data를 공유한다
즉 데이터를 공유한다는 측면은 갖지만 연산방법과 계산할수 있는 차원의 수가 다르다.
