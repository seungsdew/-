torch.cat(tensors, dim=0, out=None)
tensors: 연결할 텐서들의 시퀀스(리스트 등)입니다.
dim (옵션): 연결할 때 사용할 차원(축)을 지정합니다. 이 매개변수를 사용하여 연결할 차원을 선택할 수 있으며, 기본값은 0입니다.
out (옵션): 결과를 저장할 출력 텐서를 지정합니다. 이 매개변수를 사용하면 새로운 텐서를 생성하지 않고 결과를 저장할 수 있습니다.
예를 들어, 두 개의 텐서를 연결하려면 다음과 같이 사용할 수 있습니다:

torch.view는 모양만 바꿔주고 메모리는 똑같이 공유한다.
torch.shape는 모양만 바꿔주는 것이 아니라 메모리도 다른 공간에 할당되어 다른 텐서를 만들어 준다

>> x = torch.randn(2, 3, 5)
>>> x.size()
torch.Size([2, 3, 5])
>>> torch.permute(x, (2, 0, 1)).size()
torch.Size([5, 2, 3])
torch.permute는 차원의 순서를 서로 바꿔주고 기존의 tensor와 data를 공유한다.

torch.transpose는 2개의 차원을 transpose하고 기존의 tensor와 data를 공유한다
즉 데이터를 공유한다는 측면은 갖지만 연산방법과 계산할수 있는 차원의 수가 다르다.

```python
import torch
import matplotlib.pyplot as plt #matplotlib.pyplot를 plt로 이름만 바꿔줌
# 입력 범위 설정
x = torch.linspace(-10, 10, 100)

# 시그모이드 함수 계산
y_sigmoid = torch.sigmoid(x) #torch.linspace(start, end, steps), 시작과 끝, 그리고 몇개의 데이터를 그릴지 정하는 코드

# 시그모이드 함수의 미분 계산
y_derivative = y_sigmoid * (1 - y_sigmoid)

# 시각화
#1번 그림을 그림
plt.figure(figsize=(10, 6)) # 가로가 10이고 세로가 6인 그림을 그림
plt.subplot(2, 1, 1) # 2*1인 그림을 그리고 첫번째에 1번 그림을 그림
plt.title("Sigmoid Function") #제목을 붙임
plt.plot(x, y_sigmoid.numpy()) # 그림을 그림 (x = torch.linspace(-10, 10, 100)) ,  y는 numpy를 사용하여 텐서를 Numpy 배열형테로 만듦, 근데 없어도(y_sigmoid만 적어도)된다.
plt.grid(True) #grid 활성화

#2번 그림을 그림
plt.subplot(2, 1, 2)
plt.title("Sigmoid Derivative")
plt.plot(x, y_derivative.numpy())
plt.grid(True)

plt.tight_layout() #그림과 서브플룻의 간격을 최적화 한다. 즉 2개의 서브플룻이 겹치지 않도록 한다.
plt.show() # 그래프 표시
```

```python
import torch
import matplotlib.pyplot as plt #matplotlib.pyplot를 plt로 이름만 바꿔줌
import torch.nn
# 입력 범위 설정
x = torch.linspace(-10, 10, 100,requires_grad=True) # .backward를 사용하기 위해선 requires_grad=True를 써야 한다.

# 시그모이드 함수 계산
y_sigmoid = torch.sigmoid(x)

# 시그모이드 함수의 미분 계산
y_sigmoid.backward(torch.ones_like(x))  # 역전파 및 미분값 계산, torch.ones_like(x)를 이용해서 x와 모양만 같도록 했음

# 미분값 얻기
y_derivative = x.grad # .backward()를 사용하면 x.grad가 나온다.


# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(x.detach().numpy(), y_sigmoid.detach().numpy(), label='Sigmoid Function') #detach를 이용하여 그레디언트 추적을 비활성화 시키고 numpy와의 호환성을 유지했다.
plt.plot(x.detach().numpy(), y_derivative.detach().numpy(), label='Derivative of Sigmoid')
plt.grid(True)

plt.legend()# 범례추가
plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

class SigmoidModel(nn.Module):
    def __init__(self): #class 초기화
        super(SigmoidModel, self).__init__()

    def forward(self, x): #모델의 입력을 x로 받아서 출력을 생성
        sigmoid = torch.sigmoid(x)
        return sigmoid

def main():
    # 입력 범위 설정
    x = torch.linspace(-10, 10, 100, requires_grad=True)

    # SigmoidModel 클래스 생성
    model = SigmoidModel()

    # 모델에 입력 전달하여 시그모이드 함수 계산
    sigmoid_output = model(x)

    # 시그모이드 함수의 미분 계산
    sigmoid_output.backward(torch.ones_like(x))
    sigmoid_derivative = x.grad

    # PyTorch 텐서를 NumPy 배열로 변환
    x_numpy = x.detach().numpy()
    sigmoid_numpy = sigmoid_output.detach().numpy()
    derivative_numpy = sigmoid_derivative.detach().numpy()

    # 그래프 그리기
    plt.figure(figsize=(10, 6))
    plt.plot(x_numpy, sigmoid_numpy, label='Sigmoid Function')
    plt.plot(x_numpy, derivative_numpy, label='Derivative of Sigmoid')
    plt.grid(True)
    plt.legend()
    plt.title("Sigmoid Function and its Derivative")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__": # 스크립트가 직접 실행될 때만 main함수 실행, 다른 스크립트에서 import 되면 실행 안됨
    main()
```
